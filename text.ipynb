# **Tokenize**

```python
print("*****************************text en arabe ****************************************")
print(textall_arab)
print("----------------------------------tokenize en arabe --------------------------------------------")
textar  = nltk.tokenize.wordpunct_tokenize(textall_arab)
print(nltk.pos_tag(textar))

print("*****************************text en anglais ****************************************")
print(texten_agrs)  
print("----------------------------------tokenize en anglis --------------------------------------------")
texten  = nltk.tokenize.wordpunct_tokenize(texten_agrs)
print(nltk.pos_tag(texten))
```

##**Chunking**

```python
from nltk import pos_tag
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(pos_tag(texten))
print(result)
```

## **Stemming**
